## Visualisasi Data dengan Dist Plots

```python
#masukan modul yang digunakan 
import numpy as np
import pandas as pd 

import seaborn as sns
```

```python
#membuat angka random
data = np.random.randn(150)
```

```python
#memvisualisasikan data dengan histogram
sns.distplot(data)
```

```python
#memvisualisasikan data dengan histogram dan memberikan warna merah
sns.distplot(data, color = 'red')
```
