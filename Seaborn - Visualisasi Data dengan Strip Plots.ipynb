## Visualisasi Data dengan Strip Plots

```python
#memasukan modul yang digunakan 
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
```

```python
#memasukan data yang akan divisualisasikan
tips = sns.load_dataset('tips')
```

```python
#menampilkan 5 data pertama
tips.head()
```

```python
#memvisualisasikan tip dengan stripplot
sns.stripplot(x = tips['tip'])
```

```python
#memvisualisasikan data dengan stripplot dan memberi color = 'green'
sns.stripplot(x = tips['tip'], color = 'green')
```

```python
#memvisualisasikan total_bill dengan stripplot
sns.stripplot(x = tips['total_bill'])
```

```python
#memvisualisasikan day dan total_bill dengan stripplot
sns.stripplot(x = 'day', y = 'total_bill', data = tips)
```

```python
#memvisualisasikan day dan total_bill dengan stripplot mengunakan jitter = True
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True)
```

```python
#memvisualisasikan day dan total_bill dengan stripplot mengunakan jitter = 0.1
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = 0.1)
```

```python
#memvisualisasikan day dan total_bill dengan stripplot mengunakan jitter = 0.2
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = 0.2)
```

```python
#memvisualisasikan total_bill dan day dengan stripplot mengunakan jitter = 1.5
sns.stripplot(x = 'total_bill', y = 'day', data = tips, jitter = 1.5)
```

```python
#memvisualisasikan total_bill dan day dengan stripplot mengunakan jitter = 1
sns.stripplot(x = 'total_bill', y = 'day', data = tips, jitter = 1)
```

```python
#memvisualisasikan total_bill dan day dengan stripplot mengunakan jitter = 1 dan linewidth = 1.2
sns.stripplot(x = 'total_bill', y = 'day', data = tips, jitter = 1, linewidth = 1.2)
```

```python
#memvisualisasikan day dan total_bill berdasarkan sex
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
#memvisualisasikan day dan total_bill berdasarkan sex menggunakan jitter = True
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', jitter = True)
```

```python
#memvisualisasikan day dan total_bill berdasarkan sex menggunakan jitter True dan split = True
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', jitter = True, split = True)
```

```python
#memvisualisasikan day dan total_bill berdasarkan smoker menggunakan jitter True dan split = True
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker', jitter = True, split = True)
```

```python
#memvisualisasikan day dan total_bill berdasarkan smoker menggunakan jitter True split = True dan palette = 'winter_r'
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker', jitter = True, split = True, palette = 'winter_r')
```

```python
#memvisualisasikan day dan total_bill berdasarkan sex menggunakan jitter True split = True palette = 'winter_r'
#menggunakan order = ['Fri','Sun', 'Sat', 'Thur']
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker', jitter = True, \
              split = True, palette = 'winter_r', order = ['Fri','Sun', 'Sat', 'Thur'])
```

```python
#memvisualisasikan sex dan tip dengan stripplot
sns.stripplot(x = 'sex', y = 'tip', data = tips)
```

```python
#memvisualisasikan sex dan tip dengan stripplot berdasarkan order = ['Female', 'Male']
sns.stripplot(x = 'sex', y = 'tip', data = tips, order = ['Female', 'Male'])
```

```python
#memvisualisasikan sex dan tip dengan stripplot berdasarkan order = ['Female', 'Male'] dan palette = 'husl'
sns.stripplot(x = 'sex', y = 'tip', data = tips, order = ['Female', 'Male'], palette = 'husl')
```

```python
#memvisualisasikan sex dan tip dengan stripplot berdasarkan order = ['Female', 'Male'] palette = 'husl' dan marker = 'D'
sns.stripplot(x = 'sex', y = 'tip', data = tips, order = ['Female', 'Male'], marker = 'D', palette = 'husl')
```

```python
#memvisualisasikan day dan total_bill dengan marker = 'D'
sns.stripplot(x = 'day', y = 'total_bill', data = tips, marker = 'D')
```

```python
#memvisualisasikan day dan total_bill dengan marker = 'D' dimana size = 15
sns.stripplot(x = 'day', y = 'total_bill', data = tips, marker = 'D', size = 15)
```

```python
#memvisualisasikan day dan total_bill dengan marker = 'D' dimana size = 15 berdasarkan hue = 'sex'
sns.stripplot(x = 'day', y = 'total_bill', data = tips, marker = 'D', size = 15, hue = 'sex')
```

```python
#memvisualisasikan day dan total_bill dengan marker = 'D' dimana size = 15 berdasarkan hue = 'sex' edgecolor = gray jitter = True, alpha = 0.2 
sns.stripplot(x = 'day', y = 'total_bill', data = tips, marker = 'D', size = 15, hue = 'sex', \
              edgecolor = 'gray', alpha = 0.2, jitter = True)
```

```python
#memvisualisasikan day dan total_bill dengan marker = 'D' dimana size = 15 berdasarkan hue = 'sex' edgecolor = gray jitter = True, alpha = 0.2 dan split = True
sns.stripplot(x = 'day', y = 'total_bill', data = tips, marker = 'D', size = 15, hue = 'sex', edgecolor = 'gray', alpha = 0.2, jitter = True, split = True)
```

```python
#memvisualisasikan day dan total_bill dengan marker = 'D' dimana size = 15 berdasarkan hue = 'sex' edgecolor = gray jitter = True, alpha = 0.2 split = True dan palette = 'husl'
sns.stripplot(x = 'day', y = 'total_bill', data = tips, marker = 'D', size = 15, hue = 'sex', edgecolor = 'gray', alpha = 0.2, jitter = True, split = True, palette = 'husl')
```

```python
#memvisualisasikan day dan total_bill dengan marker = 'D' dimana size = 15 berdasarkan hue = 'sex' edgecolor = gray jitter = True, alpha = 0.5 
sns.stripplot(x = 'day', y = 'total_bill', data = tips, marker = 'D', size = 15, hue = 'sex', \
               edgecolor = 'gray', alpha = 0.5, jitter = True, split = True, palette = 'husl')
```

```python
#memvisualisasikan tip dan day dengan stripplot dan boxplot sekaligus
sns.stripplot(x = 'tip', y = 'day', data = tips)
sns.boxplot(x = 'tip', y = 'day', data = tips)
```

```python
#memvisualisasikan tip dan day dengan stripplot dan jitter = True dan boxplot sekaligus
sns.stripplot(x = 'tip', y = 'day', data = tips, jitter = True)
sns.boxplot(x = 'tip', y = 'day', data = tips)
```

```python
#memvisualisasikan tip dan day dengan stripplot dan jitter = True palette = 'winter_r' dan boxplot sekaligus
sns.stripplot(x = 'tip', y = 'day', data = tips, jitter = True, palette = 'winter_r')
sns.boxplot(x = 'tip', y = 'day', data = tips)
```

```python
#memvisualisasikan tip dan day dengan stripplot dan jitter = True palette = 'winter_r' color = '0.1 dan boxplot sekaligus
sns.stripplot(x = 'tip', y = 'day', data = tips, jitter = True, palette = 'winter_r', color = '0.1')
sns.boxplot(x = 'tip', y = 'day', data = tips)
```

```python
#memvisualisasikan tip dan day dengan stripplot dan jitter = True palette = 'winter_r' color = '1' dan boxplot sekaligus
sns.stripplot(x = 'tip', y = 'day', data = tips, jitter = True, palette = 'winter_r', color = '1')
sns.boxplot(x = 'tip', y = 'day', data = tips)
```

```python
#memvisualisasikan tip dan day dengan stripplot jitter = True dan violinplot sekaligus
sns.stripplot(x = 'tip', y = 'day', data = tips, jitter = True)
sns.violinplot(x = 'tip', y = 'day', data = tips)
```

```python
#memvisualisasikan tip dan day dengan stripplot jitter = True dan violinplot color = '0.6' sekaligus
sns.stripplot(x = 'tip', y = 'day', data = tips, jitter = True)
sns.violinplot(x = 'tip', y = 'day', data = tips, color = '0.6')
```

```python
#memvisualisasikan tip dan day dengan stripplot jitter = True dan violinplot color = '0.9' sekaligus
sns.stripplot(x = 'tip', y = 'day', data = tips, jitter = True)
sns.violinplot(x = 'tip', y = 'day', data = tips, color = '0.9')
```
