### Seaborn - Pairwise Plot

    Berikut ini adalah cara membuat pairwise plot dengan menggunakan modul seaborn.

```python
#memasukan modul yang digunakan 
import seaborn as sns
import pandas as pd
```

```python
#memasukan data
data = pd.read_excel("D:/Book1.xlsx")
data.head()
```

```python
#membuat pairwise plot
sns.set(style="ticks", color_codes=True)
g = sns.pairplot(data)
```

```python
#Plot Least Square Lines terhadap semua Prediktor
sns.pairplot(data, x_vars=['X1','X2','X3','X4'], y_vars='Y', size=7, aspect=0.7, kind='reg')
```
